```python
import os,codecs
import numpy as np
# PROVIDE YOUR DIRECTORY WITH THE EXTRACTED FILES HERE
datapath = '/Users/Fardeen/Desktop/Ece559/MNISTData/'

files = os.listdir(datapath)

def get_int(b):   # CONVERTS 4 BYTES TO A INT
    return int(codecs.encode(b, 'hex'), 16)

data_dict = {}
for file in files:
    if file.endswith('ubyte'):  # FOR ALL 'ubyte' FILES
        print('Reading ',file)
        with open (datapath+file,'rb') as f:
            data = f.read()
            type = get_int(data[:4])   # 0-3: THE MAGIC NUMBER TO WHETHER IMAGE OR LABEL
            length = get_int(data[4:8])  # 4-7: LENGTH OF THE ARRAY  (DIMENSION 0)
            if (type == 2051):
                category = 'images'
                num_rows = get_int(data[8:12])  # NUMBER OF ROWS  (DIMENSION 1)
                num_cols = get_int(data[12:16])  # NUMBER OF COLUMNS  (DIMENSION 2)
                parsed = np.frombuffer(data,dtype = np.uint8, offset = 16)  # READ THE PIXEL VALUES AS INTEGERS
                parsed = parsed.reshape(length,num_rows*num_cols)  # RESHAPE THE ARRAY AS [NO_OF_SAMPLES x HEIGHT x WIDTH]           
            elif(type == 2049):
                category = 'labels'
                parsed = np.frombuffer(data, dtype=np.uint8, offset=8) # READ THE LABEL VALUES AS INTEGERS
                parsed = parsed.reshape(length)  # RESHAPE THE ARRAY AS [NO_OF_SAMPLES]                           
            if (length==10000):
                set = 'test'
            elif (length==60000):
                set = 'train'
            data_dict[set+'_'+category] = parsed  # SAVE THE NUMPY ARRAY TO A CORRESPONDING KEY
```

```python
#init 10 vectors of 784 weights each being in the set [-10,10]
omega =  (10 - (-10)) * np.random.random_sample((10,784)) -10
print(omega.shape)
```

```python
#initializing the correct d(output)
D_train = np.zeros((60000,10))
D_test = np.zeros((10000,10))
for x in range(60000):
    row=x
    column=data_dict['train'+'_labels'][x]
    D_train[row,column] = 1
for x in range(10000):
    D_test[x,data_dict['test'+'_labels'][x]] = 1
errors =0 #thhe number of miscalculations
n=50 #the initital number of inputs =50
omega1=np.copy(omega)
epoch=0

for x in range(n):
    Y = omega1@data_dict['train'+'_images'][x,:].T  #The output vector
    myoutput = np.argmax(Y) #index of max value
    output = data_dict['train'+'_labels'][x]
    if( myoutput != output):
        errors = errors + 1
errorclassification =[epoch,errors]


#algorithm
myerrorRatio =0
errorRatio=errors/n

while errorRatio > myerrorRatio:
    epoch = epoch + 1
    errors = 0

    for i in range(n): 
        curimage = data_dict['train'+'_images'][i,:]
        D =(omega1@curimage >=0)*1
        delta =(D_train[i,:]-D)
        for x in range(10):
            omega1[x] = omega1[x] + delta[x]*curimage
        
    for x in range(n): #Determine error after one epoch
        Y = omega1@data_dict['train'+'_images'][x,:].T  #The output vector
        myoutput = np.argmax(Y) #index of max value
        output = data_dict['train'+'_labels'][x]
        if( myoutput != output):
            errors = errors + 1

    errorRatio = errors/n #refresh error rate after one epoch    
    errorclassification = np.column_stack((errorclassification,[epoch,errors]))
    print(errors)
print("epochs total: ", epoch)
print("errorRatio:",errorRatio)
        
```

```python
#To plot
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,AutoMinorLocator)

fig, ax = plt.subplots()

plt.xlabel("epoch #")
plt.ylabel("# of Errors")
plt.grid(b=True, which='both', axis='both')
plt.plot(errorclassification[0,:],errorclassification[1,:], label='Missclasification Line')

ax.legend(loc='best', shadow=True, fancybox=True, numpoints=1)
plt.show()
```

```python
#To test
errors = 0
totalTestImages = 10000
for x in range(totalTestImages):
    Y = omega1@data_dict['test'+'_images'][x,:].T  #The output vector
    myoutput = np.argmax(Y) #index of max value
    output = data_dict['test'+'_labels'][x]
    if( myoutput != output):
        errors = errors + 1
errorRatio = errors/totalTestImages
print('testing misclassifications: ', errors)
print('testing error ratio: ', errorRatio)
```
